```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv(r'E:\Projects\Final_year Project\Timebased.csv')
time_data = df["Time"]
Count_data = df["2014"]
#explode = (0.1,0,0,0,0,0,0,0)
plt.pie(Count_data,labels=time_data,autopct='%1.1f%%',shadow=True,startangle=140)
plt.title("Time Zone wise year-2014")
plt.show()
plt.savefig('plot.png',dpi=300,bbox_inches='tight')
```
